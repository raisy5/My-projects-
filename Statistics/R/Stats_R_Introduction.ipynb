```r
#5
#Calcul du diametre , circonférence/pi

diametre=x1$Girth/pi
table(diametre)

hist(diametre,xlab = "Diamètre des arbres",ylab = "Fréquence d'apparition")
```

- Le diamètre des arbres s'étend approximativement de 2,5 à 7. La distribution est **concentrée autour de 3,5-4, qui correspond à la classe la plus fréquente** (mode de lhistogramme). Cet histogramme représente la répartition des diamètres,il indique combien d'arbres appartiennent à chaque intervalle de diamètre.

---

```r
#6
#Fonction de repartition
F_repartition=ecdf(diametre)
plot(F_repartition, main = "Fonction de répartition empirique du diamètre",
     xlab = "Diamètre (pouces)", ylab = "F(x)", col = "steelblue")
```

La fonction de répartition empirique du diamètre des arbres est une fonction **croissante en escalier**, qui indique, pour chaque valeur de diamètre, la proportion d'arbres dont le diamètre est inférieur ou égal à cette valeur.

On observe une augmentation rapide de la fonction entre environ **3,5 et 4,5**, ce qui traduit **une forte concentration des diamètres dans cet intervalle**.

```r
x1 = trees
plot(x1$Girth,x1$Volume)#Affiche un nuage de point qui affiche le volume (y)
#en fonction de la circonférence (x)
pairs(x1)#Affiche un ensemble de mini graphe avec chaqu'un illustrant
# la relation entre deux variables
cor(x1)#Affiche la matrice de correlation entre les 3 variables
cor(x1,method="kendall")
cor(x1,method="spearman")
```

* Méthode Kendall

Le coefficient de Kendall mesure l'association entre deux variables en comparant toutes les paires d'observations. Une paire est dite concordante si les deux variables évoluent dans le même sens, et discordante sinon. **Le coefficient de kendall correspond à la différence (normalisée) entre le nombre de paires concordantes et discordantes**. Il varie entre -1 (relation décroissante parfaite) et +1 (croissante parfaite), et vaut 0 s'il n'y a pas de tendance monotone. C'est une méthode non paramétrique, robuste et bien **adaptée aux petits échantillons**. Elle gère généralement bien les valeurs identiques.

* Méthode Spearman

Le coefficient de Spearman mesure la relation monotone entre deux variables en utilisant leurs rangs plutôt que les valeurs brutes. Concrètement, on remplace chaque valeur par son rang, puis on calcule une corrélation (type Pearson) sur ces rangs. Il varie entre **-1 et +1, et vaut 0 en absence de relation**.Spearman est non paramétrique, donc ne suppose pas la normalité, et **il est plus robuste aux valeurs aberrantes que Pearson**. Il est très utilisé quand la relation n'est pas forcément linéaire mais reste monotone. En présence de nombreux valeurs identiques, l'interprétation reste valide mais peut être légèrement affectée.

L'espèce setosa présente des pétales nettement plus courts et très peu dispersés, indiquant une faible variabilité.

```r

par(mfrow=c(1,3), mar=c(4,4,3,1), oma=c(0,0,2,0))
hist(data$Petal.Length[iris$Species == "setosa"],main = "Setosa",xlab = "Petal.Length(cm)",
     col = "lightgray",breaks = 5)
hist(data$Petal.Length[iris$Species == "versicolor"],main = "Versicolor",xlab = "Petal.Length (cm)",
     col = "lightgray",breaks = 5)
hist(data$Petal.Length[iris$Species == "virginica"],main = "Virginica",xlab = "Petal.Length (cm)",
    col = "lightgray",breaks = 5)

par(mfrow = c(1,1))
```

- Comme on peut le constater, les données ne racontent pas la même histoire selon qu'on regroupe toutes les observations ou qu'on distingue les trois espèces. Lorsque l'on analyse la distribution des 150 fleurs sans tenir compte de l'espèce, les différences entre groupes se mélangent et peuvent donner l'illusion d'une distribution “moyenne” relativement homogène. En réalité, ce regroupement dilue l'information essentielle,chaque espèce possède sa propre distribution, avec des longueurs de pétales nettement différentes. On risque alors une interprétation erronée, par exemple conclure que les longueurs de pétales sont globalement similaires, alors que l'analyse par espèce révèle des écarts marqués.

###Paradoxe de Simpson
----
Le **paradoxe de Simpson** désigne une situation où l'analyse globale d'un jeu de données peut conduire à une conclusion trompeuse, parce que le regroupement masque  les tendances présentes dans les sous-groupes. Autrement dit, **un "effet de masse" engendré du mélange de plusieurs populations et dilue leurs différences**, ce qui fausse l'interprétation si l'on ignore les catégories qui constitue notre population. Dans notre exemple, la longueur des pétales peut sembler relativement homogène lorsqu'on considère l'ensemble des 150 observations. En revanche, dès qu'on distingue les espèces, on observe trois distributions nettement différentes ,la masse cachait donc une structure essentielle, et **l'interprétation devient totalement différente lorsqu'on analyse par catégories**.
