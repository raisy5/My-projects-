#Présentation de la Banque de Tunisie (BT)
----
![banque tunisienne](https://www.entreprises-magazine.com/wp-content/uploads/2021/12/Banque-de-Tunisie.jpg)

La Banque de Tunisie (BT), fondée en 1884, est l'une des plus anciennes institutions financières de la Tunisie et un acteur majeur du secteur bancaire tunisien. Elle opère en tant que banque universelle, offrant des services aux particuliers, entreprises et professionnels. Son capital social s'élève à 270 millions de dinars, avec un siège situé au cœur de Tunis, rue de Turquie. BT est cotée à la Bourse de Tunis et affiche une gouvernance solide, avec des partenaires internationaux comme la Banque Fédérative du Crédit Mutuel. Elle joue un rôle clé dans le financement de l'économie nationale et l'accompagnement des projets structurants.
